```python
X = df.drop("price",axis=1)
y = df["price"]
X.columns
```

```python
X.head()
```

```python
X_one_code = pd.get_dummies(X,columns=["model","transmission","fuelType"],drop_first=True)
```

```python
X_one_code = X_one_code.astype(int)
X_one_code.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
arr = ["model","transmission","fuelType"]
X_one_skl = X
for i in arr:
    X_one_skl[i] = le.fit_transform(X_one_skl[i])
X_one_skl.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
arr1 = ["year","mileage","tax","mpg"]
X_one_code[arr1] = scaler.fit_transform(X_one_code[arr1])
X_one_code.head()
```

```python
arr3 = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax']
X_one_skl[arr3] = scaler.fit_transform(X_one_skl[arr3])
X_one_skl.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_code, y, test_size=0.20, random_state=42)
```

```python
model1 = LinearRegression()
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
r2 = r2_score(y_test,y_pred)
r2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_skl, y, test_size=0.20, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(X_train,y_train)
```

```python
y_pred = model2.predict(X_test)
r2 = r2_score(y_test,y_pred)
r2
```

```python
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor()
model3.fit(X_train,y_train)
```

```python
y_pred = model3.predict(X_test)
r2 = r2_score(y_test,y_pred)
r2
```
